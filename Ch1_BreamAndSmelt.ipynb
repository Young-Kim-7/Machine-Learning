```python
fish_target = [1] * 35+ [0] *14
print(fish_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Make a object of the class 'KNeighborsClassifier'
kn = KNeighborsClassifier()
```

```python
kn.fit(fish_data, fish_target)
```

```python
kn.score(fish_data, fish_target)
```

```python
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)

for n in range(5, 50):
  # Set the number of neighbors in KNN model
  kn.n_neighbors = n

  # Calculate the score
  score = kn.score(fish_data, fish_target)

  # Print the number of neighbors who do not reach 100% accuracy
  if score < 1:
    print(n, score)
    break
```
